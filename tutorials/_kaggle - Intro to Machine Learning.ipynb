## 1. Decision trees

A simple decision tree asks if a house has more than 2 bedrooms, and whether this affects price with 'No' predicting a lower price.

This divides houses into two categories, and then the price of that group is determined by the historical average of all houses in that group.

This simple decision tree doesn't capture any other factors e.g. number of bathrooms, location etc.

More factors can be captured using a tree with more splits, called 'deeper' trees.

Each factor provides another layer of the decision tree, and the bottom layer where a prediction is made is called a 'leaf'.

## 4. Model validation

### Mean Absolute Error

In most applications, the relevant measure of model quality is predictive accuracy.

Often, the mistake is made where predictions are made using training data, and compared to real values in the training data.

First, model quality must be summarised into an interpretable metric. We will use the **Mean Absolute Error (MAE)**.

Here, the absolute value of each error i.e. $error = actual - predicted$, is calculated and the average of all absolute error values taken.

'On average, the predictions made differ from real values by X amount.'

```python
# Import data
df_melb = pd.read_csv(MELB)
# Drop NaN
df_melb = df_melb.dropna(axis = 0)
# Define features
list_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
# Define predictor
y = df_melb['Price']
# Define features
X = df_melb[list_features]

# Define model
model = DecisionTreeRegressor(random_state = 1)
# Fit data
model.fit(X, y)
# Predict values
y_pred = model.predict(X)
# Calculate MAE
mean_absolute_error(y, y_pred)
```

### Validation data

In the above example, a single sample of houses was used for building and evaluating the model.

This may accidentally capture some random association between the predicted value and some feature in the single sample, that is not generalisable .

A model's value comes from making predictions on new data, and so performance must be measured on data **not used** to build the model i.e. validation data.

```python
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state = 0)
# Define model
model = DecisionTreeRegressor()
# Fit model
model.fit(X_train, y_train)
# Generate predictions based on validation data
y_pred = model.predict(X_val)
# Calculate MAE
mean_absolute_error(y_val, y_pred)
```

This shows a massive discrepancy between the in-sample and out-sample MAE (~$430 to ~$266k).

## 5. Underfitting and overfitting

The `DecisionTreeRegressor()` model has many options, but the most important is the depth.

It is not uncommon to have trees with 10 splits between the top level and a leaf.

More splits leads to smaller numbers of houses in each leaf. These will produce accurate predictions for those houses, but may be unreliable for new data. 

This is called **overfitting* where a model matches training data closely but generalises poorly.

On the other hand, if the tree is too shallow to split data into meaningful groups, prediction accuracy will be low even within the training data, and further on new data. This is called **underfitting**.

### Controlling tree depth

`max_leaf_nodes` provides a way to control over/underfitting. The more leaves, the more likely overfitting occurs.

A utility function can be defined to compare MAE scores from different values for `max_leaf_nodes`.

```python
# Define function
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    # Define model
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    # Fit model
    model.fit(train_X, train_y)
    # Generate predictions
    preds_val = model.predict(val_X)
    # Calculate MAE
    mae = mean_absolute_error(val_y, preds_val)
    
    return mae
```

```python
# Define leaf node numbers
list_max_leaves = [5, 50, 500, 5000]

# Iterate over leaf number
for max_leaf_nodes in list_max_leaves:
    # Calculate mae
    mae = get_mae(max_leaf_nodes, X_train, X_val, y_train, y_val)
    # Report
    print(f'Max leaf nodes: {max_leaf_nodes} \t \t Mean Absolute Error: {mae}')
```

Here, 500 is shown to be the optimal leaf number, given the lowest MAE.

## 6. Random forests

Decision trees face a lot of over/underfitting problems.

Random forest uses many trees, and averages the predictions of each tree to generate a prediction. 

This generally has a much better predictive accuracy than a single tree, and works well with default parameters.

More complex models can produce better predictions, but rely heavily on parameter testing.

```python
# Define model
model = RandomForestRegressor(random_state = 1)
# Fit model
model.fit(X_train, y_train)
# Generate predictions
y_pred = model.predict(X_val)
# Calculate MAE
mean_absolute_error(y_val, y_pred)
```

This MAE improves over the single decision tree of ~$266k.
