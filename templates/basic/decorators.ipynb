# Decorators

Functional decorators extend python functions, and can help with duplication within function code.

Decorators work as a wrapper to a function, that then adds some functionality in a modular fashion.

## Decorating Simple Functions

Simple functions have no parameters.

The following code creates `decorator_function()` which:
- Takes a base function
- Defines `enhanced_function()` that adds some functionality to the base function
- Returns the enhanced function

```python
# Define decorator
def decorator_function(base_function):

    def enhanced_function():
        # Code executed before base function
        print('Before base function')
        # Base function
        base_function()
        # Code executed after base function
        print('After base function')
        print()

    return enhanced_function
```

This decorator can then be applied **when defining** any relevant simple base function.

```python
@decorator_function
def first_function():
    print('Some string')

@decorator_function
def second_function():
    print('Some other string')

@decorator_function
def third_function():
    print('Final string')

# Call functions
first_function()
second_function()
third_function()
```

## Decorating Functions w/ Parameters

Here, the decorator definition must take into account parameters of the base function.

Decorators can be modified for use with base functions that have multiple or no parameters.

- `*args` in the `enhanced_function()` definition packs all given positional arguments into a tuple.

- `*args` in the `base_function()` call unpacks the tuple into individual arguments.

- `**kwargs` in `enhanced_function()` packs all keyword arguments into a dictionary

- `**kwargs` in the `base_function()` call unpacks this dictionary

```python
# Define decorator
def decorator_function(base_function):

    def enhanced_function(*args, **kwargs):
        # Code executed before base function
        print('Before base function')
        # Base function
        base_function(*args, **kwargs)
        # Code executed after base function
        print('After base function')
        print()

    return enhanced_function
```

This now makes the decorator flexible enough to be applied to base functions with no arguments, positional arguments AND keyword arguments.

```python
# Simple function
@decorator_function
def first_function():
    print('Some string')

# Function w/ positional parameters
@decorator_function
def second_function(param1, param2):
    print(f'Parameter 1: {param1}')
    print(f'Parameter 2: {param2}')

# Function w/ keyword parameters
@decorator_function
def third_function(param1 = 'value1', param2 = 'value2'):
    print(f'Parameter 1: {param1}')
    print(f'Parameter 2: {param2}')

# Call functions
first_function()
second_function('positional1', 'positional2')
third_function(param1 = 'keyword1', param2 = 'keyword2')
```

## Handling `return` values

Most base functions will also have return values and the decorator can be extended to handle this.

```python
def decorator_function(base_function):

    def enhanced_function(*args, **kwargs):
        # Code executed before base function
        print('Before base function')
        # Base function
        result = base_function(*args, **kwargs)
        # Code executed after base functon
        print('After base function')
        print()
        # Return result
        return result
    
    return enhanced_function

@decorator_function
def first_function():
    print('Adding 1 + 2')
    result = 1 + 2
    return result

a = first_function()
print(a)
```

This syntax already handles multiple results.

```python
@decorator_function
def second_function():
    print('Adding 1 + 2')
    result1 = 1 + 2
    print('Adding 2 + 3')
    result2 = 2 + 3
    return result1, result2

a, b = second_function()
print(a, b)
```

The decorator would only need to be changed if you wanted to manipulate the multiple results **within** the enhanced function itself.

```python
def decorator_function(base_function):

    def enhanced_function(*args, **kwargs):
        # Code executed before base function
        print('Before base function')
        # Base function w/ individual results
        result1, result2 = base_function(*args, **kwargs)
        print('After base function')
        print('Adding 10 to result1')
        # Manipulate individual result(s)
        new_result1 = result1 + 10
        print()
        # Return results
        return new_result1, result2
    
    return enhanced_function

@decorator_function
def second_function():
    print('Adding 1 + 2')
    result1 = 1 + 2
    print('Adding 2 + 3')
    result2 = 2 + 3
    return result1, result2

a, b = second_function()
print(a, b)
```

**However**, this new decorator now explicitly requires functions with exactly two return results.
