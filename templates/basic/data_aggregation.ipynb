# Data Aggregation

## Example

```python
# Imports
import pandas as pd

# Define data
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [24, 30, 22, 35],
    "score": [88, 75, 90, 60],
    "group": ["A", "B", "A", "B"]
})

# Show data
df.head()
```

## `groupby`

Groups data by one or more columns.

```python
# Group data
df_grouped = df.groupby(by = 'group')[['age', 'score']]
# Perform some aggregation
df_grouped.mean()
```

## `agg`

Apply multiple aggregations after `groupby()`

```python
# Apply aggregation
df.groupby('group').agg({
    'score': ['mean', 'max', 'min'],
    'age': 'median'
})
```
