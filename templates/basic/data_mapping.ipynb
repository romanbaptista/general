# Data Mapping

## Example

```python
# Imports
import pandas as pd

# Define data
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [24, 30, 22, 35],
    "score": [88, 75, 90, 60],
    "group": ["A", "B", "A", "B"]
})

# Show data
df.head()
```

## `map`

```python
# Define mapping
group_map = {'A': 'High', 'B': 'Low'}

# Apply mapping as new column
df['group_level'] = df['group'].map(group_map)

# Show data
df.head()
```
