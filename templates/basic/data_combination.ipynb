# Data Combination

## Example

```python
# Imports
import pandas as pd

# Define data
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [24, 30, 22, 35],
    "score": [88, 75, 90, 60],
    "group": ["A", "B", "A", "B"]
})

# Show data
df.head()
```

## `merge`

Applies SQL-style joins.

```python
# Define data
df2 = pd.DataFrame({
    'group': ['A', 'B'],
    'tutor': ['Dr Smith', 'Dr Jones']
})

# Merge
df_merged = pd.merge(df, df2, on = 'group', how = 'left')

# Show data
df_merged.head()
```

## `concat`

Stacks DataFrames vertically or horizontally.

```python
# Define data
df_top = df.iloc[:2]
df_bottom = df.iloc[2:]

print(df_top)
print()
print(df_bottom)
```

### Vertical Stack

```python
# Stack data
pd.concat([df_top, df_bottom], axis = 0)
```

### Horizontal Stack

```python
# Stack data
pd.concat([df_top.reset_index(drop = True), df_bottom.reset_index(drop = True)], axis = 1)
```
