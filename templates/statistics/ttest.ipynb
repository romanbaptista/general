# T-tests

## Example

```python
# Imports
from scipy.stats import ttest_ind, ttest_rel
import numpy as np

# Data
group1 = np.array([10, 12, 9, 11, 13])
group2 = np.array([14, 15, 13, 16, 14])

t1 = np.array([50, 55, 53, 52, 54])
t2 = np.array([56, 58, 57, 55, 59])
```

## `ttest_ind`

Independent t-test, used when comparing **two separate groups** e.g. treatment vs control.

- `t_stat`: How many standard errors is the difference in means between groups from zero
- `p_value`: Probability of observing this difference if true means were equal

```python
# Run t-test
t_stat, p_value = ttest_ind(group1, group2)
# Report
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

## `ttest_rel`

Paired sample t-test, used to compare the **same subjects** before and after.
- Tests whether mean of the differences is zero
- Appropriate for repeated measures or matched pairs

```python
# Run t-test
t_stat, p_value = ttest_rel(t1, t2)
# Report
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```
